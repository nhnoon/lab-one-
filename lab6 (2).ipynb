# Linear Regression

### Scenario
I am working as a Data Analyst for a medical insurance company. The company wants to predict the **medical charges** of customers based on their personal and health-related information.  
In this lab, I will build a **Linear Regression** model using the **insurance.csv** dataset and evaluate its performance.

## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

## Check out the Data

```python
insurance = pd.read_csv('insurance.csv')
```

```python
insurance.head()
```

```python
print(insurance.columns,"\n")
print(insurance.info())
```

```python
insurance.describe()
```

## EDA
to show the trends and insights in data

```python
sns.pairplot(insurance)
plt.show()
```

```python
sns.histplot(insurance['charges'], kde=True)
plt.show()
```

```python
sns.heatmap(insurance.corr(numeric_only=True),annot=True)
plt.show()
```

```python
sns.boxplot(x='smoker', y='charges', data=insurance)
plt.show()
```

## Data Preprocessing
Converting categorical columns into numerical values using dummy variables before training the model.

```python
insurance = pd.get_dummies(insurance, drop_first=True)
insurance.head()
```

## Training a Linear Regression Model
splitting data into train, test data ....

```python
X = insurance.drop('charges',axis=1)
y = insurance['charges']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

## Model Evaluation

```python
print(lm.intercept_)
```

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

#### Interpreting the coefficients
Holding all other features fixed:

* A 1 unit increase in **age** is associated with a change in **charges** equal to the coefficient of age.
* A 1 unit increase in **bmi** is associated with a change in **charges** equal to the coefficient of bmi.
* A 1 unit increase in **children** is associated with a change in **charges** equal to the coefficient of children.
* If **smoker_yes = 1**, the expected medical charges increase significantly compared to a non-smoker.
* The variables related to **sex** and **region** show how charges may change relative to the reference category.

## Predictions from our Model

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
plt.xlabel('Y Test (True Values)')
plt.ylabel('Predicted Y')
plt.title('Actual vs Predicted Charges')
plt.show()
```

```python
sns.histplot(y_test-predictions, bins=50, kde=True)
plt.show()
```

## Regression Evaluation Metrics

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

This notebook applies **Linear Regression** to the medical insurance dataset in order to predict **charges**.  
The workflow included:
- loading and exploring the data,
- performing basic visualizations,
- converting categorical features into numerical form,
- splitting the dataset into training and testing sets,
- fitting the model,
- generating predictions,
- and evaluating the model using **MAE, MSE, and RMSE**.

The model can help the company understand which features most affect medical charges, and variables such as **smoker status** are expected to have a strong influence on the final prediction.
